```python
def QuanQual(dataset):
    Quan=[]
    Qual=[]
    for columnName in dataset.columns:
        print(columnName)
        if(dataset[columnName].dtypes=="O"):
            #print("Qual")
            Qual.append(columnName)
        else:
            #print("Quan")
            Quan.append(columnName)
    return Quan,Qual
```

```python
QuanQual(dataset)
```

```python
quan,qual=QuanQual(dataset)
```

```python
dataset[quan]
```

```python
qualdataset=dataset[qual]
```

```python
qualdataset
```

```python
#"Imputer" is used to "replace" the null value in the "multiple columns" in the dataset. (only the Quan dataset used for better result)
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(dataset[quan])
dataset=imp.transform(dataset[quan])
```

```python
imp
```

```python
dataset
```

```python
quandataset=pd.DataFrame(dataset,columns=["sl_no", "ssc_p", "hsc_p", "degree_p", "etest_p", "mba_p", "salary"])
```

```python
quandataset
```

```python
frames = [quandataset, qualdataset]

Finaldataset = pd.concat(frames,axis=1)
```
