```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.drop("User ID",axis=1)
```

```python
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
indep=dataset[["Age","EstimatedSalary","Gender_Male"]]
dep=dataset["Purchased"]
```

```python
indep.shape
```

```python
dep
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
param_grid = {
    'n_neighbors': [ 5, 10,50,100],  
    'weights': ['uniform', 'distance',None],  
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'] } 
grid = GridSearchCV(KNeighborsClassifier(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(indep, dep)
print("Best parameters found: ", grid.best_params_)
y_pred = grid.predict(indep)
print("Predictions: ", y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(dep, y_pred)
print(cm)       
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(dep, y_pred)
```

```python
print(clf_report)
```
