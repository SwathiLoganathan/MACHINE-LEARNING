```python
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups DT.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson', splitter='random',max_features=None)
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```
